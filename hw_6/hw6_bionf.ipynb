1. Создала окружение hhmer.

2. Установила в это окружение: conda install bioconda::hmmer

3. Установила тулы от ncbi, которые позволяют скачивать последовательности геномов - datasets to download biological sequence data across all domains of life from NCBI, dataformat to convert metadata from JSON Lines format to other formats.

Для этого создала и  активировала среду ncbi_datasets: conda create -n ncbi_datasets

conda activate ncbi_datasets

и установила пакеты:
conda install -c conda-forge ncbi-datasets-cli

4. Далее нашла геномы двух организмов: Drosophila melanogaster и Musca domestica.


5. Скачала белковые последовательности геномов:
datasets download genome accession GCF_030504385.1 --include protein

6. Перешла в окружение, где установлен hmmer.

7. Выбрала консервативный белок, связывающийся с TATA-box в ДНК для инициации транскрипции. image.png

8. Скачала файл формата .hmm

Запустила hmmsearch: hmmsearch PF09247.hmm ~/ncbi_dataset/data/GCF_000002985.6/protein.faa

9. Выбрала из результатов hmmsearch по белку у каждого организма с минималбьным e-value:

Musca domestica: transcription initiation factor TFIID subunit - 4.9e-19


Drosophila m.: TBP-associated factor 1, isoform A [Drosophila melanogaster] - 2.1e-18

10. Далее на сайте ncbi нашла последовательности нуклеотидные этих белков и скачала fasta.
